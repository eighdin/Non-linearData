Graph Applications

1. Shortest Path

```python
road_network_graph = nx.Graph()
road_network_graph.add_weighted_edges_from(
    [
        ('A','B',12), ('A','C',10), ('A', 'E', 15), 
        ('B','D',9), ('B','E',8), ('B','G',10), ('B','F',15),
        ('C','E',3),  
        ('D','G',5), ('D','F',6),
        ('E','F',8)
    ]
)
pos_layout = nx.circular_layout(road_network_graph)
weight_labels = []
for n1, n2, edge_weight in road_network_graph.edges(data=True):
    weight_labels.append(((n1, n2), edge_weight['weight']))
weight_labels_dict = dict(weight_labels)
#nx.draw_networkx(road_network_graph)
#road_network_graph.edges(data=True)
# nx.shortest_path(road_network_graph, 'A', 'F', weight="weight")
# nx.shortest_path(road_network_graph, 'C', 'G', weight="weight")
nx.draw_networkx(road_network_graph, pos=pos_layout)
nx.draw_networkx_edge_labels(road_network_graph, pos=pos_layout, edge_labels=weight_labels_dict, font_color="red")
```

2. Minimum Spanning Tree

```python
road_mst_graph = nx.minimum_spanning_tree(road_network_graph)
colored_graph = nx.coloring.greedy_color(road_mst_graph)
pos_mst = nx.spring_layout(road_mst_graph)
mst_weights = []
for n1, n2, node_weight in road_mst_graph.edges(data=True):
    mst_weights.append(((n1,n2), node_weight['weight']))
mst_weights_dict = dict(mst_weights)
nx.draw_networkx(road_mst_graph, pos=pos_mst)
nx.draw_networkx_edge_labels(road_mst_graph, pos=pos_mst, edge_labels=mst_weights_dict, font_color="red")
```
